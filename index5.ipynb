```python
#python এ নামের মধ্যে মান বসানোর মাধ্যমে Variable তৈরি করা হয়ে থাকে আলাদা কোনো সুস্পষ্ট নাম ঘোষনার দরকার পড়েনা।
x=2
name="Alice"
items=[2,4,6,8]
#আলাদা করে বারবার print করেও Output দেখা যায়।
print(x)
print(name)
print(items)

#একসাথে print করেও Output দেখা যায় তবে এক্ষেত্রে Comma বসাতে হবে।

print(x,name,items)

#নাম অবশ্যই একটি অক্ষর বা আন্ডারস্কোর দিয়ে শুরু করতে হবে; বাকি অংশ অক্ষর, সংখ্যা বা আন্ডারস্কোর হতে পারে। নাম কেস-সেনসিটিভ: `count` এবং `Count` ভিন্ন।

#একটি ভেরিয়েবলকে যেকোনো ধরনের নতুন মানে পুনঃনির্ধারণ করা যেতে পারে।

x=2
x="Two"
x=[2,4]
print(x)


# Multiple Assignment ২ ধরনের। যেমনঃ Chained & Parrallel

#Chained assignment এ প্রতিটি elements এক-ই Object নির্দেশ করে যার immutable values(None,integer) নিরাপদ তবে value change করা যায়না সবগুলো element এর value এক হয়।
x=y=z=None
print(x,y,z)
a=b=c=[]
a.append(2)
print(a)
print(b)
print(c)

#Parallel assignment এ প্রতিটি elements ভিন্ন ভিন্ন হতে পারে ও একলাইনে-ই ভিন্ন ভিন্ন multiple value assign করা যায়।

a,b,c=2,4,6
print(a)
print(b)
print(c)
print(a,b,c)




```
