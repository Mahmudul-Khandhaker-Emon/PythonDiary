The Print()Function

```python
print("Hello World")  ##প্রিন্ট করবে Single arguments.

print(" Hello","World","!")##প্রিন্ট করবে Multiple arguments.

print("2","4","6", sep="-")

##sep=" " দিয়ে একটা element থেকে আরেকটা element আলাদা করে।

print("Elon Musk is a Enterpreneur",end=" ")
print("and he is also a Ceo and Founder of SpaceX")

## end=" " হলো default new line print করে।


## এক লাইনে Multiple values রাখা যায় 

name = "Alice"
score = 82
print("Name:", name, "Score:", score)

## Colon(:) একটা block কে শুরু করে নিয়মানুযায়ী block শুরু করতে হলে colon(:)ব্যবহার করা আবশ্যক।

if True:
    print("Hello World!")

else: 
    print("Error")
```
